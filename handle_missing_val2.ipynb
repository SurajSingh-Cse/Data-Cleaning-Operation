```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
plt.show()
```

```python
null_var=(df.isnull().sum()/df.shape[0])*100
null_var
```

```python
drop_col=null_var[null_var >2].keys()
drop_col

```

```python
df2=df.drop(columns=drop_col)
```

```python
df2.shape
```

```python
sns.heatmap(df2.isnull())
plt.show()
```

```python
df3=df2.dropna() #it is used to remove missing value rows
df3.shape
```

```python
df3.isnull().sum()
```

```python
sns.heatmap(df3.isnull()) #dataset if clean
plt.show()
```
