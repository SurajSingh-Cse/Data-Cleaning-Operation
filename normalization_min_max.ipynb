```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("loan_data.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,4))
sns.displot(x="principal", data=df)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(df[["principal"]])
```

```python
df["principal_min_max"]=ms.transform(df[["principal"]])
```

```python
df.head(2)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title("before")
sns.histplot(x="principal",data= df)

plt.subplot(1,2,2)
plt.title("after")
sns.histplot(x="principal_min_max",data=df)
plt.show()
```
