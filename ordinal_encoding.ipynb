```python
import pandas as pd
data=pd.DataFrame({"name":["suraj","reeni","vivek",
                           "shukla","raman","prachi","oaza",
                           "kusum","mom","dad","jiju","dadji"]})
data
```

```python
ord_data=[["suraj","reeni","vivek","shukla","raman","prachi",
           "oaza","kusum","mom","dad","jiju","dadji"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(data[["name"]])
```

```python
data["name_en"]=oe.transform(data[["name"]])
data
```

```python
ord_data1={"suraj":1,"reeni":2,"vivek":3,"shukla":4,"raman":5,"prachi":6,
           "oaza":7,"kusum":8,"mom":9,"dad":10,"jiju":11,"dadji":12}
```

```python
data["name_en_map"]=data["name"].map(ord_data1)
```
