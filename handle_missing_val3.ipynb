handle missing value and fill catagorical dataset

```python
#import library

import pandas as pd # type: ignore
import seaborn as sns # type: ignore
import matplotlib.pyplot as plt # type: ignore
 

 #load dataset
df=pd.read_csv(r"loan_data.csv")
df.head(4)

```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.fillna(10) #fill 10 in missing content
```

```python
df.info()
```

```python
df.ffill (axis=1) #forward filling
```

```python
df.bfill() #backward fill
```

```python
df.select_dtypes(include="object").isnull().sum()
```

```python
df.select_dtypes(include="object").columns #it is give me catagorical data columns
```

```python
for i in df.select_dtypes(include="object").columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```
