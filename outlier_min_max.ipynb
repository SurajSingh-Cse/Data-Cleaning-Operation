```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("loan_data.csv")
df.head(4)

```

```python
df.isnull().sum()
```

```python
sns.boxenplot(x="principal",data=df)
plt.show()
```

```python
sns.histplot(x="principal",data=df)
plt.show()
```

```python
q1=df["principal"].quantile(0.25)
q3=df["principal"].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_range=q1 -(1.5*IQR)
max_range=q3 +(1.5*IQR)
min_range, max_range
```

```python
new_df = df[df["principal"]<=max_range]
```

```python
new_df.shape
```

```python
sns.boxenplot(x="income",data=new_df)
plt.show()
```
